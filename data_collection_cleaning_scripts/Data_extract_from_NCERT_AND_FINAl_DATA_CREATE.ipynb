# KEST

```python
soup = BeautifulSoup(open("kest109_glossary_only.html"), 'html.parser')
```

```python
all_terms_texts = [i.text.strip() for i in soup.find_all('p', {"class": ["ft0", "ft2"]})][2:]
all_terms_texts
```

```python
all_terms = []
all_texts = []
for trm_txt in all_terms_texts:
  trm = ''
  txt = ''
  tkns = trm_txt.split(' ')
  for i in range(len(tkns)):
    if tkns[i][0].isupper() and tkns[i+1][0].isupper():
      trm = trm + ' ' + tkns[i]
    else:#if tkns[i][0].isupper() and tkns[i+1][0].islower():
      txt = ' '.join(tkns[i:]) 
      break
  all_terms.append(trm.strip())
  all_texts.append(txt.strip())
```

```python
all_terms
```

```python
all_texts
```

```python
ncert_data = pd.DataFrame({'terms':all_terms,'definitions':all_texts})
ncert_data['source'] = 'ncert_kest'
ncert_data['assigned_readibility']=1.0
ncert_data
```

```python
ncert_data.to_csv('ncert_data_kest.csv', index = False)
```

# NCERT Class - 12

```python
soup = BeautifulSoup(open("NCERT Class 12 Economics_glossary.html"), 'html.parser')
```

```python
all_terms = [i.text.strip() for i in soup.find_all('span', {"class": ["ft0", "ft2"]})]
all_terms
```

```python
all_terms_corrected = []
for i in range(len(all_terms)-1):
  curr_trm = all_terms[i]
  nxt_trm = all_terms[i+1]
  if curr_trm[0].isupper() and nxt_trm[0].islower():
    all_terms_corrected.append(curr_trm + ' ' + nxt_trm)
  else:
    all_terms_corrected.append(curr_trm)
all_terms_corrected    
```

```python
len(all_terms_corrected)
```

```python
all_texts = [i.text.strip() for i in soup.find_all('p')]
all_texts
```

```python
len(all_texts)
```

```python
terms_li=[]
texts_li = []
for term in all_terms:
  for text in all_texts:
    if term in text:
      terms_li.append(term)
      texts_li.append(text.replace(term,'').strip())
```

```python
ncert_data = pd.DataFrame({'terms':terms_li,'definitions':texts_li})
ncert_data['source'] = 'ncert_class12_econ'
ncert_data['assigned_readibility']=1.0
ncert_data
```

```python
ncert_data.to_csv('ncert_data_class12_econ.csv', index = False)
```

```python
import textstat
```

```python
def extract_scores(test_data):
  try:
    scores = [textstat.flesch_reading_ease(test_data),textstat.flesch_kincaid_grade(test_data),textstat.smog_index(test_data),textstat.coleman_liau_index(test_data), textstat.automated_readability_index(test_data),textstat.dale_chall_readability_score(test_data), textstat.difficult_words(test_data), textstat.linsear_write_formula(test_data), textstat.gunning_fog(test_data), textstat.text_standard(test_data), textstat.fernandez_huerta(test_data), textstat.szigriszt_pazos(test_data), textstat.gutierrez_polini(test_data), textstat.crawford(test_data), textstat.gulpease_index(test_data), textstat.osman(test_data)]
  except:
    scores = [0.0]*9 + ['NOT USE'] + [0.0]*6
  return scores
```

```python
scores = new_data['definitions'].apply(lambda x: extract_scores(str(x)))
scores_df = pd.DataFrame(list(scores))
scores_df.columns = ['flesch_reading_ease','flesch_kincaid_grade','smog_index','coleman_liau_index','automated_readability_index','dale_chall_readability_score','difficult_words','linsear_write_formula','gunning_fog','text_standard','fernandez_huerta','szigriszt_pazos','gutierrez_polini','crawford','gulpease_index','osman']
scores_df
```

```python
new_data_scores_df = pd.concat([new_data.reset_index(),scores_df.reset_index()], axis = 1).drop('index', axis = 1)
new_data_scores_df
```
